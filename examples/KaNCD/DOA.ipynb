```python
import pandas as pd
import json
from collections import defaultdict



# 处理数据
import ast

# 如果 knowledge_code 是字符串，需要转换为列表
data_DOA["knowledge_code"] = data_DOA["knowledge_code"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

result = defaultdict(lambda: defaultdict(list))

for _, row in data_DOA.iterrows():
    for code in row["knowledge_code"]:
        # 按照 knowledge_code 和 item_id 分组
        result[code][row["item_id"]].append({
            "user_id": row["student_id"],
            "score": row["correct"],
        })

# 构造目标 JSON 格式
json_result = []
for knowledge_code, exercises in result.items():
    exercises_list = []
    for exer_id, logs in exercises.items():
        exercises_list.append({
            "exer_id": exer_id,
            "logs": logs
        })
    json_result.append({
        "knowledge_code": knowledge_code,
        "exercises": exercises_list
    })

# 输出结果
with open("output.json", "w", encoding="utf-8") as f:
    json.dump(json_result, f, ensure_ascii=False, indent=4)

print("JSON 数据已保存到 output.json 文件中")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class CustomModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(CustomModel, self).__init__()
        self.input_dim = input_dim
        self.output_dim = output_dim
        # 定义卷积层，降维到 output_dim
        self.conv = nn.Conv1d(
            in_channels=1, 
            out_channels=output_dim, 
            kernel_size=1
        )

    def forward(self, x, input_knowledge):
        # 输入形状: [batch_size, input_dim]
        batch_size, input_dim = x.shape
        assert input_dim == self.input_dim, f"Expected input_dim={self.input_dim}, got {input_dim}"

        # Step 1: 降维
        x = x.unsqueeze(1)  # 添加通道维度，形状 [batch_size, 1, input_dim]
        x = self.conv(x)  # 卷积，形状 [batch_size, output_dim, input_dim]
        x = x.squeeze(2)  # 去掉多余维度，形状 [batch_size, output_dim]

        assert x.shape[1] == self.output_dim, (
            f"Expected output_dim={self.output_dim}, got {x.shape[1]}"
        )

        # Step 2: Softmax
        x_softmax = F.softmax(x, dim=-1)  # [batch_size, output_dim]

        # Step 3: 找出每行最大值的两个维度
        topk_values, topk_indices = torch.topk(x_softmax, 2, dim=-1)  # [batch_size, 2]

        # 确保 topk_indices 不超出范围
        assert torch.max(topk_indices) < self.output_dim, (
            f"Index out of bounds: max(topk_indices)={torch.max(topk_indices)}, "
            f"output_dim={self.output_dim}"
        )

        mask = torch.zeros_like(x_softmax)  # 初始化掩码
        mask.scatter_(1, topk_indices, 1)  # 在掩码中设置前两大值的位置为 1

        # Step 4: 合并 softmax mask 和 input_knowledge
        assert input_knowledge.shape == mask.shape, (
            f"input_knowledge shape {input_knowledge.shape} does not match mask shape {mask.shape}"
        )
        merged = torch.logical_or(mask.bool(), input_knowledge.bool()).float()  # 合并掩码

        # Step 5: 对合并后的结果进行归一化
        normalized = F.normalize(merged, p=2, dim=1)  # L2 范数归一化

        return normalized


# 测试代码
batch_size = 32
input_dim = 4096  # 输入维度
output_dim = 100  # 降维后的 k
input_tensor = torch.randn(batch_size, input_dim)  # 输入向量 [batch_size, input_dim]
input_knowledge = torch.randint(0, 2, (batch_size, output_dim)).float()  # 模拟 one-hot 向量 [batch_size, output_dim]

model = CustomModel(input_dim, output_dim)
output = model(input_tensor, input_knowledge)

print("Output shape:", output.shape)  # 应为 [batch_size, output_dim]
print(output)

```
