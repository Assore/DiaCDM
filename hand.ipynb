```python
import random
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

def generate_handwritten_text(text, output_file, font_paths, image_size=(200, 100), text_color="black"):
    # Create a blank image with white background
    img = Image.new("RGB", image_size, "white")
    draw = ImageDraw.Draw(img)

    # Randomly select a font from the given paths and a random font size
    font_path = random.choice(font_paths)
    font_size = random.randint(20, 40)
    font = ImageFont.truetype(font_path, font_size)

    # Calculate text size using textbbox (for newer versions of Pillow)
    text_bbox = draw.textbbox((0, 0), text, font=font)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]

    # Center the text within the image dimensions
    position = ((image_size[0] - text_width) // 2, (image_size[1] - text_height) // 2)

    # Draw the text
    draw.text(position, text, fill=text_color, font=font)
    img.save(output_file)
    return img

# Your font paths
font_paths = ["C:/Windows/Fonts/FZYTK.ttf", "C:/Windows/Fonts/FZSTK.ttf", "C:/Windows/Fonts/SIMYOU.ttf"]

# Generate and display multiple images
generated_images = []
for i in range(5):  # Generate 5 variations
    output_file = f"handwritten_{i}.png"
    img = generate_handwritten_text("贾瑞", output_file, font_paths)
    generated_images.append(img)

# Display the generated images
fig, axs = plt.subplots(1, len(generated_images), figsize=(15, 5))
for ax, img in zip(axs, generated_images):
    ax.imshow(img)
    ax.axis("off")
plt.show()
```
